# **Graph Knowledge: nhận diện danh từ riêng**
---

* **A	- Tính từ (mạnh, dứt khoát, tích cực,...)**
* C	- Quan hệ từ: từ dùng để biểu thị ý nghĩa quan hệ như so sánh, sở hữu, nhân quả (và,...)
* CH - Dấu câu
* E	- Chỉ từ: từ dùng để trỏ vào sự vật nhằm xác định vị trí của sự vật trong không gian, thời gian. Ví dụ: này, kia, ấy, nọ
* L	- Lượng từ: từ chỉ lượng ít hay nhiều của sự vật… VD: Tất cả, mọi, những, mấy, vài, dăm, từng, các, mỗi..
* **M	- Từ chỉ số lượng: 100, 1000,...**
* **N	- Danh từ**
* **Np - Danh từ riêng (tên riêng: HPG, Vnindex, anh Thìn,...)**
* Nu	- Tiền tệ (VND, USD)
* P	- Tình thái từ: từ dùng để thêm vào câu để tạo nên câu nghi vấn, câu cầu khiến, câu cảm thán hoặc biểu thị sắc thái tình cảm của người nói. Ví dụ: à, ư, hử, nhỉ, chăng, chứ (nghi vấn) đi, nào, với (cầu khiến) thay, sao... (cảm thán) ạ, nhé, cơ mà... (sắc thái tình cảm)
* R	- Phó từ: từ chuyên đi kèm với động từ, tính từ để bổ sung ý nghĩa cho nó. VD: đã-đang-sẽ, rất-lắm-quá, cũng-từng, không-chưa-chẳng, được…
* **V	- Động từ (mua, bán, chốt, duy trì, bứt phá,...)**
* X	3- Quán ngữ (nói tóm lại, đáng chú ý là, mặt khác thì…)
